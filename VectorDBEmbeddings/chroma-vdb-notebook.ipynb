```python
# set up a client
import chromadb
client = chromadb.Client()
```

```python
# create a 'collection' lik in mongodb
collection1 = client.create_collection(name="my_collection1")
collection2 = client.create_collection(name="my_collection2")
```

```python
collection1.add(
    documents=["This is a document", "This is another document"],
    metadatas=[
        {
            "source": "source1",
            "language": "en"
        },
        {
            "source": "source2"
        }
    ],
    ids=[ "id1", "id2" ]
)
```

```python
#methods ca be run on clients
client.list_collections()
```

```python
# Get an existing collection
collection1 = client.get_collection("my_collection1")
```

```python
# Get a collectin or create if it doest not exist

collection3 = client.get_or_create_collection("my_collection3")
```

```python
# delete a collection
client.delete_collection("my_collection3")
```

```python
# utility methods
# heartbeat
client.heartbeat()
```

```python
# to reset collection. can't be undone
# client.reset()
```

```python
# count # of items in a collection
collection1.count()
```

```python
# adding new item to existing collection
collection1.add(
    documents=["This is document 3", "This is document 3"],
    metadatas=[{"source": "source3", "language": "en"}, {"source": "source4"}],
    ids=[ "id3", "id4" ]
)
```

```python
collection1.get()
```

```python
collection1.upsert(
    documents=["doc3", "This is document 4"],
    metadatas=[{"source": "source3", "language": "en"}, {"source": "source4"}],
    ids=[ "id3", "id4" ]
)
```

```python
collection1.get()
```
